```python
###########################
# Look Test Data
###########################
import pandas as pd

df = pd.read_csv('./Test.csv',header= 0)
df.head(15)
```

```python
##############################
# Generate Test Data
##############################
import re
import glob 
import os
import shutil

taille = len(df['Path'])
mycolumns = ['ClassId','Path']
liste = []
tupl = ()
for i in range(taille):
        p = df['Path'][i]
        c = df['ClassId'][i]
        p = re.sub('.*\/', '', p)
        tupl = (int(c),p)
        liste.append(tupl)

for filepath in glob.iglob('./Test/*.png'):
        filepath = re.sub('.*\/', '', filepath)
        for x in liste:         
                if filepath in x:
                        if not os.path.exists(f'./Test/{str(x[0])}'):
                                path = os.path.join('./Test/',str(x[0]))
                                os.mkdir(path)
                        shutil.move(f'./Test/{x[1]}',f'./Test/{str(x[0])}')
```
